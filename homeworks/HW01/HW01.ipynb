```python
# 1
import numpy as np
```

```python
# 2
# из обычного списка
arr_example = [1, 2, 3.4, 5]

numpy_arr_1 = np.array(arr_example)

numpy_arr_1
```

```python
#  с использованием функций `zeros`, `ones`, `arange` или аналогичных.

numpy_arr_2 = np.zeros([10,4]) # матрица из нулей, 10 строк, 4 столбца
numpy_arr_3 = np.ones([3, 6]) # матрица из единиц, 3 строки, 6 столбцов
numpy_arr_4 = np.arange(10) # вектор из чисел от 0 до 10
```

```python
numpy_arr_2
```

```python
numpy_arr_4
```

```python
# 3
numpy_arr_3.shape # размер массива, 3 строки, 6 столбцов
```

```python
numpy_arr_4.dtype # тип данных в массиве, int32
```

```python
# 4 индексирование и срезы

numpy_arr_3[0, 1] # получение отдельного элемента, 1 строка, 2 столбец
```

```python
numpy_arr_4[4:6] # подмассив, от 5 элемента до 7, не включая последний
```

```python
# 5
example_1 = np.arange(5)
example_2 = np.arange(5)

sum_arrays = example_1 + example_2 # поэлементное сложение
sum_arrays
```

```python
mul_array = example_1 * example_2 # поэлементное умножение
mul_array
```

```python
# вычисление средней величины (mean) или другой простой статистики
mean = np.mean(sum_arrays) # средняя величина
mean
```

```python
# сумма
summary = np.sum(sum_arrays)
summary
```

```python
median = np.median(sum_arrays) # медиана
median
```

```python
std = round(np.std(sum_arrays), 3) # с.к. отклонение
std
```

```python
# максимум и минимум
maximum = np.max(sum_arrays)
minimum = np.min(sum_arrays)

maximum, minimum
```
